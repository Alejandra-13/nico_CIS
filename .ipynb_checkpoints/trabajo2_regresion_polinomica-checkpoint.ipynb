```python
##importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import PolynomialFeatures 
```

```python
#importamos los datos
data1 = pd.read_csv('data/datos_polinomio.csv',delimiter=',')
```

```python
#Organizamos los datos
X1 = data1.iloc[:,4]
X1= X1[:, np.newaxis]
Y1= data1.iloc[:,5]
Y1 = Y1[:, np.newaxis]
print("dimension en X {}" .format(X1.shape))
print("dimension en Y {}" .format(Y1.shape))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X1,Y1,test_size=.3, random_state=3)
```

```python
#graficamos
plt.scatter(X1,Y1,color='blue', marker='o',alpha=0.5)
plt.title("Gráfica 1")
plt.xlabel("Años de experiencia")
plt.ylabel("Salario")
plt.show()
```

```python
f, [ax1, ax2] = plt.subplots(2, 1, sharey=True) 
ax1.scatter(X_train,Y_train,color='blue', marker='^',alpha=0.5)
ax2.scatter(X_test, Y_test ,color='blue', marker='^',alpha=0.5)
plt.show()
```

```python
regresion = LinearRegression()
regresion.fit(X_train,Y_train)
```

```python
plt.scatter(X_train, Y_train, color="red", marker='^', alpha=0.5)
plt.plot(X_train, regresion.predict(X_train), color="blue")
plt.title("Sueldo vs años de experiencia - Conjunto de entrenamiento")
plt.xlabel ("Años de experiencia")
plt.ylabel ("Sueldo $")
plt.show()
```

```python
pf = PolynomialFeatures(degree = 8)    # usaremos polinomios de grado 3
X_poli = pf.fit_transform(X_train.reshape(-1,1))  # transformamos la entrada en polinómica
regresion_lineal = LinearRegression()
regresion_lineal.fit(X_poli, Y_train)
```

```python
plt.scatter(X_train, Y_train, color="red", marker='^', alpha=0.5)
plt.scatter(X_train, regresion_lineal.predict(X_poli), marker='.', color="blue")#probar con plot
plt.title("Sueldo vs años de experiencia - Conjunto de entrenamiento")
plt.xlabel ("Años de experiencia")
plt.ylabel ("Sueldo $")
plt.show()
```

```python
X_train_organizado=np.sort(X_train[:,0])
Y_predecido_organizado=np.sort(regresion_lineal.predict(X_poli)[:,0])
plt.plot(X_train_organizado, Y_predecido_organizado, marker='.', color="blue")#probar con plot
```

```python
# datos_organizados_2=pd.DataFrame([X_train[:,0],regresion_lineal.predict(X_poli)[:,0]]).transpose()
datos_organizados_2=pd.DataFrame([X_train[:,0],regresion_lineal.predict(X_poli)[:,0]]).transpose()
datos_organizados_2.columns=['x_organizado_2','y_organizado_2']
datos_organizados_2=datos_organizados_2.sort_values('x_organizado_2')
X1_organizado = datos_organizados_2.iloc[:,0]
X1_organizado= X1_organizado[:, np.newaxis]
Y1_organizado = datos_organizados_2.iloc[:,1]
Y1_organizado= Y1_organizado[:, np.newaxis]
plt.plot(X1_organizado, Y1_organizado, marker='.', color="blue")#probar con plot
```

```python
plt.scatter(X_train, Y_train, color="red", marker='^', alpha=0.5)
plt.plot(X1_organizado, Y1_organizado, color="blue")#probar con plot
plt.title("Sueldo vs años de experiencia - Conjunto de entrenamiento")
plt.xlabel ("Años de experiencia")
plt.ylabel ("Sueldo $")
plt.show()
```
