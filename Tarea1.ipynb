* Función que genera en cada casion un a serie de puntos aleatorios

```
def valores():
  df=pd.DataFrame(
    {'x': range(1,21),
     'y': np.random.randn(20), 
     'z': np.random.randn(20)
     })
  return(df)

def dibujar(graf):
  fig, axs = plt.subplots( len(graf), 1)
  for x in range(len(graf)):
    axs[x].grid(True);axs[x].plot(graf[x].x,graf[x].y);axs[x].set_ylabel('varaible en y{}'.format(x+1))  
  plt.xlabel('variable en x')  
  plt.show()
```

```
graf=[valores(),valores(),valores()]
dibujar(graf)
```
